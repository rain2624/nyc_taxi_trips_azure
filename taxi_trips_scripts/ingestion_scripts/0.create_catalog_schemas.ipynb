## Create External Location, Catalog and Schemas required for the project
- Create catalog.
- Creating schema - bronze, silver and gold (with managed location).

### 1. Creating Catalog

```sql
CREATE CATALOG IF NOT EXISTS taxi_trips_2025;
```

```sql
USE CATALOG taxi_trips_2025;
```

### 2. Creating external location 
(before this I have created storage credentials using UI).

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `taxi_trips_bronze_layer`
URL 'abfss://bronze@nyctaxitrips2025.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `taxi-trips-storage-creds`);
```

```sql
DESC EXTERNAL LOCATION taxi_trips_bronze_layer
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `taxi_trips_silver_layer`
URL 'abfss://silver@nyctaxitrips2025.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `taxi-trips-storage-creds`);
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `taxi_trips_gold_layer`
URL 'abfss://gold@nyctaxitrips2025.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `taxi-trips-storage-creds`);
```

### 3. Creating Schema

```sql
CREATE SCHEMA IF NOT EXISTS bronze
MANAGED LOCATION "abfss://bronze@nyctaxitrips2025.dfs.core.windows.net"
```

```sql
CREATE SCHEMA IF NOT EXISTS silver
MANAGED LOCATION "abfss://silver@nyctaxitrips2025.dfs.core.windows.net"
```

```sql
CREATE SCHEMA IF NOT EXISTS gold
MANAGED LOCATION "abfss://gold@nyctaxitrips2025.dfs.core.windows.net"
```
